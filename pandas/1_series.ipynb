```python
import pandas as pd 

#create series 
s1 = pd.Series([1,2,3,4,5]) #no labels so index starting from 0 will be label 
print(s1)
```

```python
list = [10,20,30,'car',50.99]
labels = ['a','b','c','d','e']
#create Series with list and labels 
s2 = pd.Series(list,labels)
print(s2)
print("data type of Series ",s2.dtype)

```

```python
cities = {'bhavnagar':42.11,'surat':43.25,'dwarka':38.20}
#create Series from dictionary
s3 = pd.Series(cities) #here key will be label and value will be value 
print(s3)
```

```python
import numpy as np 

#create an array 
a1 = np.array([1,2,3,4,5])

#create series from array 
s4 = pd.Series(a1)
print(s4)
```
